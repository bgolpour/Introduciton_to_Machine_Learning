```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mglearn
```

```python
#We assume that we have many different ways to observe the
# mixture (say 100 measurement devices), each of which provides us with a series of
# measurements:
# mix data into a 100-dimensional state
A = np.random.RandomState(0).uniform(size=(100, 3))
X = np.dot(S, A.T)
print("Shape of measurements: {}".format(X.shape))
```

```python
from sklearn.decomposition import NMF
from sklearn.decomposition import PCA
```

```python
S = mglearn.datasets.make_signals()
plt.figure(figsize=(6, 1))
plt.plot(S, '-')
plt.xlabel("Time")
plt.ylabel("Signal")
```

```python
nmf = NMF(n_components=3, random_state=42)
S_ = nmf.fit_transform(X)
print("Recovered signal shape: {}".format(S_.shape))
```

```python
# For comparison, we also apply PCA:
pca = PCA(n_components=3)
H = pca.fit_transform(X)
```

```python
# We need to show the signal activity that was discovered by NMF and PCA:
models = [X, S, S_, H]
names = ['Observations (first three measurements)',
        'True sources',
        'NMF recovered signals',
        'PCA recovered signals']
fig, axes = plt.subplots(4, figsize=(8, 4), gridspec_kw={'hspace': .5},
                        subplot_kw={'xticks': (), 'yticks': ()})
for model, name, ax in zip(models, names, axes):
    ax.set_title(name)
    ax.plot(model[:, :3], '-')
```
