```python
from scipy import sparse
eye = np.eye(4)
print("NumPy array:\n{}".format(eye))
```

```python
#convert the NumPy to csr SciPy sparse matrix in CSR format
sparse_matrix = sparse.csr_matrix(eye)
print("\nSciPy sparse CSR matrix:\n{}".format(sparse_matrix))
```

```python
data =  np.ones(10)
row_indices = np.arange(10)
col_indices = np.arange(10)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print("COO representation:\n{}".format(eye_coo))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape:{}".format(y_train.shape))
```

```python
# Create a dataframe
# Lable the columns using the strings in iris_dataset.feature_names
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
```

```python
# Create a scatter matrix from the dataframe, color by y_train
from pandas.plotting import scatter_matrix
scatter_matrix(iris_dataframe, alpha=0.2, figsize=(10, 10))
```

```python
pd.plotting.scatter_matrix(iris_dataframe, alpha=0.2, figsize=(10, 10))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Prediction:{}".format(prediction))
print("Prediction target name:{}".format(
        iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Test set prediction:\n{}".format(y_pred))
```

```python
print("Test set score:{:.2f}".format(np.mean(y_pred == y_test)))
```

```python
test_set_accuracy = format(knn.score(X_test, y_test))
print(test_set_accuracy)
```

```python
##OR...
print("Test set score: {:.5f}".format(knn.score(X_test, y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train,y_train)
```

```python
print("Test set prediction: {}".format(clf.predict(X_test)))
```

```python
print("Test set accuracy: {:.2f}".format(clf.score(X_test,y_test)))
```
